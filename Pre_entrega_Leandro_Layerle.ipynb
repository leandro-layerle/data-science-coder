# **Funcion para evitar repetir el codigo de la lectura del archivo**

```python
def ReadGitHubUrl(url):
  try:
    dataframe = pd.read_csv(url)
    return dataframe
  except Exception as e:
    print(f"Error al leer el archivo CSV desde la URL. Detalles: {e}")
    return None
```

# **Leo el archivo desde GitHub**

```python
df = ReadGitHubUrl('https://raw.githubusercontent.com/leandro-layerle/data-science-datasets/main/spotify_songs.csv')
df.head()
```

# **Análisis Exploratorio de Datos (EDA)**

**Visualizaciones y Análisis:**

Histograma de la popularidad de las canciones.

Gráfico de dispersión entre popularidad y danceability.

Gráfico de barras para la distribución de géneros musicales.

## **1. Histograma de la popularidad de las canciones:**


```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
plt.hist(df['track_popularity'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribución de Popularidad de Canciones', fontsize=18)
plt.xlabel('Popularidad', fontsize=14)
plt.ylabel('Frecuencia', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True)
plt.show()
```

## **2. Gráfico de dispersión entre popularidad y danceability:**


```python
plt.figure(figsize=(12, 8))
plt.scatter(df['track_popularity'], df['danceability'], c='blue', alpha=0.5)
plt.title('Relación entre Popularidad y Danceability', fontsize=18)
plt.xlabel('Popularidad', fontsize=14)
plt.ylabel('Danceability', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True)
plt.show()
```

## **3. Gráfico de barras para la distribución de géneros musicales:**


```python
plt.figure(figsize=(12, 8))
genre_counts = df['playlist_genre'].value_counts()
genre_counts.plot(kind='bar', color=['skyblue', 'orange', 'green', 'red', 'purple'])
plt.title('Distribución de Géneros Musicales', fontsize=18)
plt.xlabel('Género Musical', fontsize=14)
plt.ylabel('Número de Canciones', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y')
plt.show()
```

## **¿Cuáles son las características musicales más comunes entre las canciones populares?**

```python
import matplotlib.pyplot as plt

# Calcular las características musicales promedio para las canciones populares
caracteristicas_promedio = df.groupby('track_popularity').mean()

# Seleccionar las características relevantes (por ejemplo, danceability, energy, valence)
caracteristicas_relevantes = caracteristicas_promedio[['danceability', 'energy', 'valence']]

# Colores para las líneas
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']

# Graficar las características promedio
plt.figure(figsize=(12, 8))
for i, col in enumerate(caracteristicas_relevantes.columns):
    plt.plot(caracteristicas_relevantes.index, caracteristicas_relevantes[col], label=col, color=colors[i], linewidth=2)

plt.title('Características Musicales Promedio de Canciones Populares', fontsize=16)
plt.xlabel('Popularidad de la Canción', fontsize=14)
plt.ylabel('Valor Promedio', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

## **¿Existe una relación entre la popularidad y la danceability de las canciones?**


```python
# Graficar la relación entre popularidad y danzabilidad
plt.figure(figsize=(12, 8))
plt.scatter(df['track_popularity'], df['danceability'], color='salmon', alpha=0.5, s=50)
plt.title('Relación entre Popularidad y Danceability', fontsize=16)
plt.xlabel('Popularidad de la Canción', fontsize=14)
plt.ylabel('Danzabilidad', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

## **¿Cómo se distribuyen las canciones en cuanto a géneros en el conjunto de datos de Spotify?**


```python
import matplotlib.pyplot as plt

# Contar la cantidad de canciones por género
canciones_por_genero = df['playlist_genre'].value_counts()

# Colores para el gráfico de pastel
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0', '#ffb3e6', '#c2f0c2']

# Crear el gráfico de pastel
plt.figure(figsize=(10, 8))
patches, texts, autotexts = plt.pie(canciones_por_genero, labels=canciones_por_genero.index, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Distribución de Canciones por Género', fontsize=18)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()

# Aumentar el tamaño de fuente de los números internos del gráfico
for autotext in autotexts:
    autotext.set_fontsize(12)

# Aumentar el tamaño de fuente de los nombres de los géneros
for text in texts:
    text.set_fontsize(12)

plt.show()
```

# **Analisis de Metadata**

### **Para Variables Numéricas:**


*Análisis de otras variables numéricas (danceability, energy, valence, tempo, duration_ms):*


```python
# Resumen estadístico de variables numéricas
numeric_summary = df[['danceability', 'energy', 'valence', 'tempo', 'duration_ms']].describe()
print("Resumen estadístico de variables numéricas:")
print(numeric_summary)

# Histogramas
plt.figure(figsize=(12, 10))
for i, col in enumerate(['danceability', 'energy', 'valence', 'tempo', 'duration_ms'], 1):
    plt.subplot(3, 2, i)
    plt.hist(df[col], bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Distribución de {col}')
    plt.xlabel(col)
    plt.ylabel('Frecuencia')
plt.tight_layout()
plt.show()
```

## **Para Variables Categóricas:**


*1. Análisis de 'playlist_genre':*

```python
# Total de canciones
total_canciones = len(df['track_id'])
print("Número total de canciones en el dataset:", total_canciones)

# Número de canciones por género
genre_counts = df['playlist_genre'].value_counts()
print("\nNúmero de canciones por género:")
print(genre_counts)

# Proporción de canciones en cada género
genre_proportions = genre_counts / len(df)
print("\nProporción de canciones en cada género:")
print(genre_proportions)
```

*2. Análisis de otras columnas*


```python
# Número de canciones por cada artista
artist_counts = df['track_artist'].value_counts()

# Los 5 principales artistas
top_artists = artist_counts.head(5)
print("Los 10 principales artistas:")
print(top_artists)

# Gráfico de barras
plt.figure(figsize=(12, 6))
top_artists.plot(kind='bar', color='skyblue')
plt.title('Los 10 principales Artistas')
plt.xlabel('Artista')
plt.ylabel('Número de Canciones')
plt.xticks(rotation=45)
plt.show()

# El artista mas escuchado
artista_mas_escuchado = df['track_artist'].value_counts().idxmax()
print("\nEl artista más escuchado es:", artista_mas_escuchado)

# Duración promedio de las canciones
duracion_promedio = df['duration_ms'].mean() / (1000 * 60) # Convertir milisegundos a minutos
print("\nDuración promedio de las canciones (minutos):", duracion_promedio)

# Popularidad promedio
popularidad_promedio = df['track_popularity'].mean()
print("\nPopularidad promedio de las canciones:", popularidad_promedio)

# Año de lanzamiento más común
año_mas_comun = df['track_album_release_date'].str[:4].mode()[0]
print("\nAño de lanzamiento más común:", año_mas_comun)

# Número total de álbumes
total_albumes = df['track_album_name'].nunique()
print("\nNúmero total de álbumes:", total_albumes)

# Popularidad mínima y máxima
popularidad_minima = df['track_popularity'].min()
popularidad_maxima = df['track_popularity'].max()
print("\nPopularidad mínima de las canciones:", popularidad_minima)
print("Popularidad máxima de las canciones:", popularidad_maxima)

# Duración mínima y máxima de las canciones
duracion_minima = df['duration_ms'].min() / (1000 * 60) # Convertir milisegundos a minutos
duracion_maxima = df['duration_ms'].max() / (1000 * 60) # Convertir milisegundos a minutos
print("\nDuración mínima de las canciones (minutos):", duracion_minima)
print("Duración máxima de las canciones (minutos):", duracion_maxima)

# Número de playlists
total_playlists = df['playlist_name'].nunique()
print("\nNúmero total de playlists:", total_playlists)
```
